```python
#Importamos Librerías
import pandas as pd     
from datetime import datetime 
```

```python
# Diagnósticos asignados por el hospital, codificados mediante el International Statistical Classification of Diseases DNA Related Health Problems (ICD)
DIAGNOSES_ICD= pd.read_csv('PF_UCI-/datasets/DIAGNOSES_ICD.csv')

# Grupos relacionados por el diagnóstico, que el hospital utiliza para facturación
DRGCODES= pd.read_csv('PF_UCI-/datasets/DRGCODES.csv')

# Exámenes de laboratorio para pacientes del hospital y para consultas externas
LABEVENTS= pd.read_csv('PF_UCI-/datasets/LABEVENTS.csv')

# Medidas y estudios microbiológicos de la base de datos del hospital
MICROBIOLOGYEVENTS= pd.read_csv('PF_UCI-/datasets/MICROBIOLOGYEVENTS.csv')

# Órdenes de medicamentos, no siempre administradas, para un paciente
PRESCRIPTIONS= pd.read_csv('PF_UCI-/datasets/PRESCRIPTIONS.csv')
```

DIAGNOSES_ICD

```python
# Extraemos los nombres ombres de columnas en una lista
DIAGNOSES_ICD_NAMES = list(DIAGNOSES_ICD.columns.values)
# Descripcion de columnas
DIAGNOSES_ICD_DESCRIPCIÓN = ['Identificador de fila único','ID del Paciente','Id de Admisiones','Prioridad del código. La secuencia 1 es el código principal','Clasificación Internacional de Enfermedades']
# Unimos la lista con nombres de columnas y la descripcion de cada columna
DIAGNOSES_ICD_ND= dict(zip(DIAGNOSES_ICD_NAMES,DIAGNOSES_ICD_DESCRIPCIÓN))
# Creamos el df
DIAGNOSES_ICD_columnas = pd.DataFrame([[key, DIAGNOSES_ICD_ND[key]] for key in DIAGNOSES_ICD_ND.keys()], columns=['Name', 'Descripcion'])
# Tipo de dato de cada columna
DIAGNOSES_ICD_dtype = ['INT','INT','INT','INT','VARCHAR(10)']
# Agregamos el tipo de dato
DIAGNOSES_ICD_columnas['Dtype'] = DIAGNOSES_ICD_dtype
# --------------------
# PK
DIAGNOSES_ICD_PK=['Y','','','','']
# FK
DIAGNOSES_ICD_FK=['','Y','Y','','Y']
# Agregamos el PK
DIAGNOSES_ICD_columnas['PK']=DIAGNOSES_ICD_PK
# Agregamos el FK
DIAGNOSES_ICD_columnas['FK']=DIAGNOSES_ICD_FK
# -------------------------
# Agregamos nombre de columna
DIAGNOSES_ICD_columnas['tabla'] = 'DIAGNOSES_ICD'
# Agregamos número de columna
DIAGNOSES_ICD_columnas['n tabla'] = 16
#Print el resultado
DIAGNOSES_ICD_columnas
```
